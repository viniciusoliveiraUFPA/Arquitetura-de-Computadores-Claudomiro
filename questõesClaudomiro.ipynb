# Questão 1

*Descreva as funções básicas de uma UCP, indicando os seus componentes principais.*

Resposta: A UCP (também chamada de CPU — Central Processing Unit) é o componente responsável por executar as instruções de um programa, controlando e coordenando todas as operações realizadas pelo computador.
A UCP pode ser dividida em três grandes blocos:

- Unidade de Controle (UC)
Responsável por coordenar e comandar todas as operações do processador.
Gera os sinais de controle que indicam a cada componente o que fazer e quando fazer (com base no clock).
Interpreta as instruções (decodificação) e direciona o fluxo de dados entre os demais componentes.
- Unidade Lógica e Aritmética (ULA / ALU)
Executa as operações aritméticas (soma, subtração, multiplicação, etc.) e lógicas (AND, OR, NOT, comparações, deslocamentos de bits, etc.).
Recebe os operandos vindos dos registradores e devolve o resultado, que pode gerar também flags (bits de estado, como overflow, carry, zero, etc.).
- Registradores: Pequenas unidades de memória de altíssima velocidade, internas à UCP, usadas para armazenar dados temporários e informações de controle. Os principais são:

  - Contador de Programa (PC – Program Counter): guarda o endereço da próxima instrução a ser buscada.
  - Registrador de Instrução (RI – Instruction Register): armazena a instrução atualmente em execução.
  - Acumulador (ACC): armazena resultados intermediários de operações da ULA (em arquiteturas mais simples).
  - Registrador de Endereço de Memória (REM/MAR): guarda o endereço de memória a ser acessado.
  - Registrador de Dados de Memória (RDM/MBR): guarda o dado lido ou a ser escrito na memória.
  - Registrador de Estado/Flags: armazena os bits indicadores de estado da última operação da ULA.

# Questão 2

*Quais são as funções da unidade aritmética e lógica – UAL?*

Resposta: A Unidade Aritmética e Lógica é o componente da UCP responsável por executar efetivamente as operações de processamento de dados. Suas funções principais são:
- Operações aritméticas: soma, subtração, multiplicação, divisão, incremento/decremento e comparação de valores.

- Operações lógicas: AND, OR, NOT, XOR, deslocamentos (shifts) e rotações de bits.

- Geração de flags (bits de estado), usadas depois pela Unidade de Controle para desvios condicionais: Z (zero), C (carry), V (overflow), N (negativo)

- Fluxo: recebe operandos + código da operação (vindo da UC) → produz resultado + flags.

# Questão 3

*O que é e para que servia o ACC?*

Resposta: O Acumulador (ACC) é um registrador especial dentro da UCP, usado como ponto central de armazenamento temporário dos resultados das operações realizadas pela ULA. Em arquiteturas mais simples e antigas (conhecidas como arquiteturas baseadas em acumulador), o ACC funcionava como um dos operandos implícitos e o destino padrão das operações aritméticas e lógicas:

1. Um dos operandos de uma operação já estava sempre no ACC.
2. O outro operando vinha de outro registrador ou diretamente da memória.
3. A ULA realizava a operação.
4. O resultado era automaticamente armazenado de volta no próprio ACC, sobrescrevendo o valor anterior.

# Questão 5

*Quais são as funções da unidade de controle de um processador?*

Resposta: A Unidade de Controle é o componente da UCP responsável por coordenar e comandar todas as operações do processador, sem executar diretamente os cálculos. Suas funções principais são:

1. Busca de instruções (Fetch): Comanda a busca da próxima instrução na memória, usando o endereço indicado pelo Contador de Programa (PC), e transfere essa instrução para o Registrador de Instrução (RI).

2. Decodificação (Decode): Interpreta a instrução armazenada no RI, identificando qual operação deve ser realizada, quais são os operandos envolvidos e qual o modo de endereçamento utilizado.

3. Geração de sinais de controle: Com base na instrução decodificada, a UC gera os sinais de controle que comandam os demais componentes do sistema, indicando, por exemplo:
- à ULA, qual operação executar;
- aos registradores, quando ler ou escrever dados;
- à memória, quando realizar uma leitura ou escrita;
- aos barramentos, qual caminho de dados habilitar.

4. Sequenciamento das operações: Garante que cada etapa do ciclo de instrução (busca → decodificação → execução → armazenamento) ocorra na ordem correta e no momento correto, sincronizando tudo por meio do sinal de clock.

5. Controle de fluxo do programa: Gerencia desvios (jumps) e desvios condicionais, atualizando o PC de acordo com o fluxo sequencial normal (PC = PC + 1), ou saltos determinados por instruções de desvio, muitas vezes usando as flags geradas pela ULA.

6. Coordenação da comunicação com memória e E/S: Controla o acesso do processador à memória principal e aos dispositivos de entrada/saída, gerenciando os barramentos de endereço, dados e controle para que as informações trafeguem corretamente entre eles.

# Questão 6

*Seria possível realizar o projeto de um processador em que o tamanho em bits do CI fosse diferente do tamanho do REM? Nesse caso, qual dos dois registradores deveria ter maior tamanho? Por quê?*

Resposta: Sim, é possível projetar um processador em que o Contador de Instrução (CI/PC) e o Registrador de Endereço de Memória (REM/MAR) tenham tamanhos diferentes. O REM deve ter tamanho maior ou igual ao do CI, pois O REM precisa cobrir todo o espaço de memória, incluindo a região onde estão as instruções e as regiões de dados. Já o CI só precisa cobrir a região de instruções. Logo, o conjunto de endereços que o CI pode gerar é, no máximo, igual ao que o REM cobre, nunca maior.

# Questão 7

*Considere um computador cuja MP é organizada com N células de 1 byte cada uma. As instruções interpretadas pela UCP possuem três tamanhos diferentes: as do tipo A possuem 16 bits; as do tipo B têm 32 bits e as do tipo C possuem 48 bits. Considerando que o código de operação de cada uma tem um tamanho fixo e igual a 8 bits e que os programas executados nesse processador são constituídos de uma mistura dos três tipos de instrução, imagine um processo prático para incremento automático do CI após a execução de cada instrução de um programa.*

Resposta: Como o código de operação (opcode) tem tamanho fixo de 8 bits (1 byte) e fica sempre na mesma posição (o primeiro byte da instrução), ele pode ser usado para determinar o tamanho total da instrução antes mesmo de ela ser completamente buscada e decodificada.

Processo passo a passo:

1. Busca do primeiro byte: A UC usa o valor atual do CI para buscar apenas o primeiro byte da instrução (o opcode) na memória, colocando-o no REM e depois no RDM/RI.

2. Identificação do tipo de instrução: A UC consulta esse opcode — por exemplo, através de uma tabela ou de uma lógica de decodificação — e determina a qual dos três tipos a instrução pertence (A, B ou C)

3. Busca dos bytes restantes: Com o tamanho da instrução já conhecido, a UC busca os bytes restantes (1, 3 ou 5 bytes, completando 2, 4 ou 6 bytes no total), incrementando o CI a cada byte lido, ou lendo-os de uma vez se o barramento permitir.

4. Incremento final do CI: Após a instrução estar completamente buscada, o CI já está apontando (ou é ajustado para apontar) para o endereço do byte seguinte ao último byte da instrução atual, no número de bytes correspondente ao tamanho da instrução que acabou de ser executada

# Questão 8

*Considere um processador cujo ciclo de instrução não possua a etapa de incremento automático do valor do CI. Imagine um método alternativo que permita a execução do programa.*

Resposta: Sem o incremento automático do CI (PC) durante o ciclo de busca, o processador não tem como saber qual é a próxima instrução, então essa responsabilidade precisa ser transferida para outro lugar. Uma das alternativas é, em cada instrução, além do OpCode e dos operandos, ter o endereço da instrução seguinte. Ao final da execução, o processador carrega esse campo no CI.

# Questão 11

*Qual é o registrador cujo conteúdo determina a capacidade de memória de um computador? Justifique.*

Resposta: É o REM (Registrador de Endereço de Memória), também chamado de MAR (Memory Address Register). Ele é o registrador responsável por armazenar o endereço de qualquer posição de memória que a UCP deseja acessar (para leitura ou escrita, de instrução ou de dado). É justamente o tamanho em bits desse registrador que determina quantos endereços distintos o processador consegue gerar e quanta memória ele consegue efetivamente endereçar. Essa relação é dada por 2^n, onde n é o número de bits do REM.

# Questão 15

*Quando se fala que um determinado microcomputador A é um micro de 8 bits e que um outro micro B é de 16 bits, a que estamos nos referindo? Ao tamanho da célula de MP ou ao tamanho da palavra? Qual a base desses dois conceitos (palavra e célula)?*

Resposta: Quando dizemos que um microcomputador é "de 8 bits" ou "de 16 bits", estamos nos referindo ao tamanho da palavra (word size) do processador. A célula da memória principal tem, por convenção quase universal, tamanho fixo de 1 byte (8 bits), independentemente de o processador ser de 8, 16, 32 ou 64 bits. Isso porque o byte é a unidade padrão de endereçamento da memória — cada endereço aponta para 1 byte. A palavra (word) é a unidade natural de dados que a UCP consegue processar de uma só vez — ou seja, é o tamanho, em bits, dos principais componentes internos do processador, como os registradores internos (acumulador, registradores de uso geral), a ULA (a largura dos operandos que ela processa em uma única operação), e também o barramento de dados externo.

# Questão 18

*Explique a diferença entre um processamento sequencial e um outro pipeline.*

Resposta: No processamento sequencial, cada instrução passa por todas as etapas do ciclo de instrução (busca, decodificação, execução, armazenamento) antes de a próxima instrução começar a ser processada. Ou seja, enquanto uma instrução não terminar completamente, a outra instrução nem começa a ser buscada. Os recursos do processador (UC, ULA, registradores) ficam ociosos boa parte do tempo, já que cada etapa usa apenas uma parte do hardware, mas as demais partes ficam paradas esperando.

No processamento em pipeline, as diferentes etapas do ciclo de instrução são executadas por unidades de hardware distintas e especializadas, permitindo que várias instruções estejam em estágios diferentes simultaneamente. Enquanto uma instrução está sendo executada, a outra instrução já está sendo decodificada, e outra instrução já está sendo buscada, tudo no mesmo ciclo de clock.
